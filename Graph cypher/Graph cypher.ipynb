```python
import math
import secrets
import networkx as nx
import random

#encryption
word = input("Enter your text: ") #insert your word/phrase here

n = tuple((ord(x)-96) for x in word)
# print(n)

a = tuple(x+len(word) for x in n)
# print(a)

def findbs(a_list, rand_ceil=10**2, increment=1.1):
    b_list = [0] * len(a_list)
    for ind, a in enumerate(a_list):
        while math.gcd(a, b_list[ind]) != 1:
            try:
                b_list[ind] = secrets.choice(range(max(max(b_list) +1 , 26), 
                                               int(rand_ceil)))
            except IndexError:                
                rand_ceil = rand_ceil * increment
                b_list[ind] = secrets.choice(range(max(max(b_list) +1 , 26), 
                                               int(rand_ceil))) 
    return b_list

b_list = findbs(a)
# print(b_list)

c = tuple(pow(a, -1, mod=b_list[ind]) for ind, a in enumerate(a))
# print(c)

pendulum_vertices = dict(zip(b_list, c))


G = nx.Graph()

G.add_nodes_from(random.sample(b_list, k=len(b_list)))
G.add_edges_from((node, list(G.nodes)[ind+1]) if ind < (len(G.nodes)-1) else (node, list(G.nodes)[0]) 
                 for ind, node in enumerate(list(G.nodes)))

G.add_nodes_from(c)
G.add_edges_from((b, c) for b, c in pendulum_vertices.items())

fig = nx.draw_networkx(G, with_labels=True)

#decryption
b_list.sort()
a = tuple(pow(c, -1, mod=b_list[ind]) for ind, c in enumerate(c))
#print(a)

a_sub = tuple([x - len(a) for x in a])
# a - shift value

for x in a_sub:
    chr(ord('`')+x)    
#print(str(chr(ord('`')+x)))
```
